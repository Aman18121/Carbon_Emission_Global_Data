```python
import pandas as pd
import matplotlib.pyplot as plt
import time 
```

```python
df = pd.read_csv("global_emissions.csv")
df
```

```python
def country_population():
    country = input("Enter country name")
    country_data = df[df["Country.Name"] == country.title()]

    plt.plot(country_data["Year"],country_data["Country.Population"])
    plt.xlabel("Year")
    plt.ylabel("Population")
    plt.title("Population of {} Over Years".format(country))
    plt.grid(True)
    plt.xticks(df["Year"] , df["Year"], rotation = 45)
    plt.show()

    plt.plot(country_data["Year"],country_data["Emissions.Production.CO2.Total"], color = "red")
    plt.xlabel("Year")
    plt.ylabel("Emissions.Production.CO2.Total")
    plt.title("Emissions.Production.CO2.Total of {} Over Years".format(country))
    plt.grid(True)
    plt.xticks(df["Year"] , df["Year"], rotation = 45)
    plt.show()

start_1 = time.time() * 1000
country_population()
stop_1 = time.time() * 1000
print(f"Running time for country_population: {stop_1 - start_1:.2f} ms" )
```

```python
#in depth relations of co2 emission from diffrent industries and comparisions  
def comparison ():
    country_of_intrest = input("Enter the country of intrest :").title()
    country_of_comparision = input("Enter the country you want to compare with:").title()
    if country_of_intrest not in df["Country.Name"].values or country_of_comparision not in df["Country.Name"].values:
        return("The name of the country is incorrect or not present in the data sheet")

    
    country_data = df[df["Country.Name"] == country_of_intrest.title()]
    country_data2 = df[df["Country.Name"] == country_of_comparision.title()]
    plt.plot(country_data["Year"],country_data["Country.Population"] , color = "red", label = country_of_intrest)
    plt.plot(country_data2["Year"], country_data2["Country.Population"], color = "green" , label = country_of_comparision)
    plt.xlabel("Population")
    plt.ylabel("Years")
    plt.title("Population of {} and {} Over Years".format(country_of_intrest, country_of_comparision))
    plt.grid(True)
    plt.legend()
    plt.xticks(country_data["Year"], rotation=45)
    plt.show()

    plt.figure(figsize=(6, 10))
    plt.barh(country_data["Year"], country_data["Country.Population"], color="red", alpha=0.5, label=country_of_intrest)
    plt.barh(country_data["Year"], country_data2["Country.Population"], color="green", alpha=0.5, label=country_of_comparision)
    plt.xlabel("Year")
    plt.ylabel("Population")
    plt.title("Population of {} and {} Over Years".format(country_of_intrest, country_of_comparision))
    plt.grid(True)
    plt.yticks(country_data2["Year"], rotation = 45)
    plt.legend()
    plt.show()

comparison()
```

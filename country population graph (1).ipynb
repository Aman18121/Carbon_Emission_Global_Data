```python
import pandas as pd
import matplotlib.pyplot as plt
import time 
```

```python
df = pd.read_csv("global_emissions.csv")
df
```

```python
def country_population():
    country = input("Enter country name")
    country_data = df[df["Country.Name"] == country.title()]

    plt.plot(country_data["Year"],country_data["Country.Population"])
    plt.xlabel("Year")
    plt.ylabel("Population")
    plt.title("Population of {} Over Years".format(country))
    plt.grid(True)
    plt.xticks(df["Year"] , df["Year"], rotation = 45)
    plt.show()

    plt.plot(country_data["Year"],country_data["Emissions.Production.CO2.Total"], color = "red")
    plt.xlabel("Year")
    plt.ylabel("Emissions.Production.CO2.Total")
    plt.title("Emissions.Production.CO2.Total of {} Over Years".format(country))
    plt.grid(True)
    plt.xticks(df["Year"] , df["Year"], rotation = 45)
    plt.show()

start_1 = time.time() * 1000
country_population()
stop_1 = time.time() * 1000
print(f"Running time for country_population: {stop_1 - start_1:.2f} ms" )
```

```python






#in depth relations of co2 emission from diffrent industries and comparisions  
def comparison ():
    country_of_intrest = input("Enter the country of intrest :").title()
    country_of_comparision = input("Enter the country you want to compare with:").title()
    if country_of_intrest not in df["Country.Name"].values:
        return("The name of the country you are interested in is incorrect or not present in the data sheet")
    if country_of_comparision not in df["Country.Name"].values:
        return("The name of the country you want to compare is incorrect or not present in the data sheet")
    
    country_data = df[df["Country.Name"] == country_of_intrest.title()]
    country_data2 = df[df["Country.Name"] == country_of_comparision.title()]
    plt.plot(country_data["Year"],country_data["Country.Population"] , color = "red", label = country_of_intrest)
    plt.plot(country_data2["Year"], country_data2["Country.Population"], color = "green" , label = country_of_comparision)
    plt.xlabel("Population")
    plt.ylabel("Years")
    plt.title("Population of {} and {} Over Years".format(country_of_intrest, country_of_comparision))
    plt.grid(True)
    plt.legend()
    plt.xticks(country_data["Year"], rotation=45)
    plt.show()

    plt.figure(figsize=(6, 10))
    plt.barh(country_data["Year"], country_data["Country.Population"], color="red", alpha=0.5, label=country_of_intrest)
    plt.barh(country_data["Year"], country_data2["Country.Population"], color="green", alpha=0.5, label=country_of_comparision)
    plt.xlabel("population")
    plt.ylabel("Year")
    plt.title("Population of {} and {} Over Years".format(country_of_intrest, country_of_comparision))
    plt.grid(True)
    plt.yticks(country_data2["Year"], rotation = 45)
    plt.legend()
    plt.show()

    #plt.figure(figsize=(6, 10))
    plt.scatter(country_data["Year"], country_data["Emissions.Production.CO2.Total"], color ="red", label = country_of_intrest,  marker = "s")
    plt.plot(country_data["Year"], country_data["Emissions.Production.CO2.Total"], color ="red")
    plt.scatter(country_data["Year"], country_data2["Emissions.Production.CO2.Total"], color ="green", label = country_of_comparision, marker = "o")
    plt.plot(country_data["Year"], country_data2["Emissions.Production.CO2.Total"], color ="green")
    plt.xticks(country_data2["Year"], rotation = 45)
    plt.legend()
    plt.xlabel("Year")
    plt.ylabel("Co2 emmision in million metric tons")
    plt.grid(True)
    plt.title
    plt.show 

    plt.figure(figsize=(10, 10))
    plt.barh(country_data["Year"], country_data["Country.GDP"], color="red", alpha=0.5, label=country_of_intrest)
    plt.barh(country_data["Year"], country_data2["Country.GDP"], color="green", alpha=0.5, label=country_of_comparision)
    plt.xlabel("GDP")
    plt.ylabel("Year")
    plt.title("GDP of {} and {} Over Years".format(country_of_intrest, country_of_comparision))
    plt.grid(True)
    plt.yticks(country_data2["Year"], rotation = 45)
    plt.legend()
    plt.show()

    fig = plt.figure(constrained_layout = True)
    fig.set_size_inches(12,8)
    gs =fig.add_gridspec(3 ,3)

    f_ax1 = fig.add_subplot(gs[0,:])
    f_ax1.set_title("Emissions.Production.CO2.Oil")
    f_ax2 = fig.add_subplot(gs[1,0:2])
    f_ax2.set_title("Emissions.Production.CO2.Flaring")
    f_ax3 = fig.add_subplot(gs[1:,2])
    f_ax3.set_title("Emissions.Global Share.CO2.Total")
    f_ax4 = fig.add_subplot(gs[2, 0])
    f_ax4.set_title("Emissions.Production.CO2.Gas")
    f_ax5 = fig.add_subplot(gs[2, 1])
    f_ax5.set_title("Emissions.Production.CO2.Coal")

    f_ax5.plot(country_data["Year"], country_data["Emissions.Production.CO2.Coal"], color = "red", label = country_of_intrest)
    f_ax5.plot(country_data["Year"], country_data2["Emissions.Production.CO2.Coal"], color = "green", label = country_of_comparision)
    f_ax5.legend()
    f_ax5.set_xlabel("Year")
    f_ax5.set_ylabel("Co2 emmision in million metric tons")
    f_ax5.grid(True)

    f_ax4.plot(country_data["Year"], country_data["Emissions.Production.CO2.Gas"], color = "red", label = country_of_intrest)
    f_ax4.plot(country_data["Year"], country_data2["Emissions.Production.CO2.Gas"], color = "green", label = country_of_comparision)
    f_ax4.legend()
    f_ax4.set_xlabel("Year")
    f_ax4.set_ylabel("Co2 emmision in million metric tons")
    f_ax4.grid(True)

    f_ax1.plot(country_data["Year"], country_data["Emissions.Production.CO2.Oil"], color = "red", label = country_of_intrest)
    f_ax1.plot(country_data["Year"], country_data2["Emissions.Production.CO2.Oil"], color = "green", label = country_of_comparision)
    f_ax1.legend()
    f_ax1.set_xlabel("Year")
    f_ax1.set_ylabel("Co2 emmision in million metric tons")
    f_ax1.grid(True)

    f_ax3.bar(country_data["Year"], country_data["Emissions.Global Share.CO2.Total"], color="red", alpha=0.5, label=country_of_intrest)
    f_ax3.bar(country_data["Year"], country_data2["Emissions.Global Share.CO2.Total"], color="green", alpha=0.5, label=country_of_comparision)
    f_ax3.grid(True)
    f_ax3.set_xlabel("Year")
    f_ax3.set_ylabel("Co2 emmision in million metric tons")
    f_ax3.legend()

    f_ax2.bar(country_data["Year"], country_data["Emissions.Production.CO2.Flaring"], color="red", alpha=0.5, label=country_of_intrest)
    f_ax2.bar(country_data["Year"], country_data2["Emissions.Production.CO2.Flaring"], color="green", alpha=0.5, label=country_of_comparision)
    f_ax2.grid(True)
    f_ax2.set_xlabel("Year")
    f_ax2.set_ylabel("Co2 emmision in million metric tons")
    f_ax2.legend()
    
comparison()
```
